```python
import pandas as pd
df = pd.read_csv('matches.csv')
```

```python
df.loc[:]
```

```python
wins = df['winner'].value_counts()
wins
```

```python
top10 = wins.head(10)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.bar(top10.index, top10.values)
plt.xlabel("Teams")
plt.ylabel("Wins")
plt.title("Top 10 IPL Teams by Wins")
plt.xticks(rotation=45)
plt.show()
```

```python
mvp = df['player_of_match'].value_counts().head(10)
mvp
```

```python
plt.figure(figsize=(8,6))
plt.barh(mvp.index, mvp.values)
plt.xlabel("Times")
plt.ylabel("Players")
plt.title("Top 10 IPL Player of Match Awards")
plt.show()
```

```python
nof = df['season'].value_counts().sort_index()
nof
```

```python
plt.figure(figsize=(10,5))
plt.plot(nof.index, nof.values, marker='o')
plt.xlabel('Season')
plt.ylabel('Number of matches')
plt.title('Number of matches each season')
plt.grid(True)
plt.show()
```

```python
Top_venue = df['venue'].value_counts().head(10)
plt.figure()
plt.barh(Top_venue.index, Top_venue.values)
plt.xlabel("Times")
plt.ylabel("Venue")
plt.title("Top Most Venue")
plt.show()
```

```python
df['toss_match_win'] = df['toss_winner'] == df['winner']
tosswins = df['toss_match_win'].value_counts()
plt.figure()
plt.bar(tosswins.index.astype(str), tosswins.values)
plt.xlabel("Result")
plt.ylabel("Times")
plt.title("Toss wins == match win? ")
plt.show()
```
